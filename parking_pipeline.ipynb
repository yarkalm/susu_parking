```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('frame.jpg') 
print(img.shape)
h, w = img.shape[0], img.shape[1]
print(w)
fig = plt.figure(figsize=(20, 8))
img = img[int(h*0.495):int(h*0.875), int(w*0.15):int(w*0.85)]
ax = fig.add_subplot()
ax.imshow(img)
 
plt.show()
```
